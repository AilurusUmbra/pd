# CSV File Preprocess
* modify each field to (rowName, colName, fieldValue)

```python
import numpy as np
import pandas as pd

df = pd.read_csv('NHC_PCET-Hemi_2D_scan_DFT_relative_E.csv', index_col=0)
df.head()
```

```python
for row in range(df.shape[0]):
    for col in range(df.shape[1]):        
        df.iloc[row, col] = str(df.index[row]) + ", " + str(df.columns[col]) + ", " + str(df.iloc[row, col])
```

```python
df
```

```python
df.to_csv('out.csv', sep=',')
```
